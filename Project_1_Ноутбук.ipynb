# Преобразование данных

1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
# функция для извлечения уровня образования
def education(level_university):
    level_education = level_university.split()
    level_education_two_words = " ".join(level_education[0:2]).lower()
    if level_education_two_words == 'среднее специальное' or level_education_two_words == 'неоконченное высшее':
        return level_education_two_words
    else:
        x = level_education[0].lower()
        return x

# создаем новый параметр и применяем функцию к столбцу "Образование и ВУЗ"
df_head_hunter['Образование'] = df_head_hunter['Образование и ВУЗ'].apply(education)

# удаляем параметр 'Образование и ВУЗ'
df_head_hunter.drop('Образование и ВУЗ', axis=1, inplace=True)
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:

* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
# функция для разделения столбца "Пол, возраст" на параметры:
def sex_age(sex_age_birthday):
    sex, age, _ = sex_age_birthday.split(' , ')
    sex = sex[0]
    age = int(age.split(' ')[1])
    return sex, age

# создаем новые столбцы и заполняем данными из столбца "Пол, возраст"
df_head_hunter['Пол'], df_head_hunter['Возраст'] = zip(*df_head_hunter['Пол, возраст'].apply(sex_age))

# удаляем больше не нужный столбец "Пол, возраст"
df_head_hunter.drop(['Пол, возраст'], axis=1, inplace=True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# функция по заполнению новых столбцов из столбца "Опыт работы"
def job_skill(job):
    if job is np.nan:
        return None
    if job == 'Не указано':
        return None
    job_mounth = job.split()
    if job_mounth[5] in ['месяц', 'месяца', 'месяцев']:
        return int(job_mounth[2]) * 12 + int(job_mounth[4])
    if job_mounth[3] not in ['лет', 'года', 'год']:
        return int(job_mounth[2])
    return int(job_mounth[2]) * 12

# Создание нового столбца за счет применения функции к столбцу "Опыт работы"
df_head_hunter['Опыт работы(месяц)'] = df_head_hunter['Опыт работы'].apply(job_skill)

# удаляем больше не нужный столбец "Опыт работы"
df_head_hunter.drop("Опыт работы", axis=1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# список городов-миллионников
million_cities = ['Новосибирск', 
                  'Екатеринбург',
                  'Нижний Новгород',
                  'Казань', 
                  'Челябинск',
                  'Омск', 
                  'Самара', 
                  'Ростов-на-Дону', 
                  'Уфа', 
                  'Красноярск', 
                  'Пермь', 
                  'Воронеж',
                  'Волгоград']

# функция для извлечения данных из столбца "Город, переезд, комндировки" в новые столбцы
def town_moving_trips(info):
    if len(info.split(' , ')) == 3:
        town, moving, trips = info.split(' , ')
    elif len(info.split(' , ')) == 4:
        town, _, moving, trips = info.split(' , ')
    else:
        town, moving, *trips = info.split(' , ')

    town = str(town)
    if town in million_cities:
        town = 'город-миллионник'
    elif town == 'Москва' or town == 'Санкт-Петербург':
        pass
    else:
        town = 'другие'

    if 'не ' in moving:
        moving = False
    else:
        moving = True
    if 'не ' in trips:
        trips = False
    else:
        trips = True

    return town, moving, trips

# присваиваем результат работы функции в качестве значений новых столбцов
df_head_hunter['Город'], df_head_hunter['Готовность к переезду'], df_head_hunter['Готовность к командировкам'] = zip(*df_head_hunter['Город, переезд, командировки'].apply(town_moving_trips))

# удаляем ненужный более столбец
df_head_hunter.drop(['Город, переезд, командировки'], axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# функция по заполнению новых столбов из признака "Занятость"
def emoloyment(employ):
    full, part_time, project, volonteer, intern = False, False, False, False, False
    if 'полная занятость' in str(employ):
        full = True
    if 'частичная занятость' in employ:
        part_time = True
    if 'проектная работа' in employ:
        project = True
    if 'волонтерство' in employ:
        volonteer = True
    if 'стажировка' in employ:
        intern = True
    return full, part_time, project, volonteer, intern

# применение функции по заполнению новых столбцов из признака "Занятость"
df_head_hunter['полная занятость'], df_head_hunter['частичная занятость'], df_head_hunter['проектная работа'], df_head_hunter['волонтерство'], df_head_hunter['стажировка'] = zip(*df_head_hunter['Занятость'].apply(emoloyment))

# функция по заполению новых столбцов из признака "График"
def schedule(schedule_value):
    full_day, chanqe_day, flexible_day, distant, shift_work = False, False, False, False, False
    if 'полный день' in schedule_value:
        full_day = True
    if 'сменный график' in schedule_value:
        chanqe_day = True
    if 'гибкий график' in schedule_value:
        flexible_day = True
    if 'удаленная работа' in schedule_value:
        distant = True
    if 'вахтовый метод' in schedule_value:
        shift_work = True
    return full_day, chanqe_day, flexible_day, distant, shift_work

# применение функции по заполнению новых столбцов из признака "График"
df_head_hunter['полный день'], df_head_hunter['сменный график'], df_head_hunter['гибкий график'], df_head_hunter['удаленная работа'], df_head_hunter['вахтовый метод'] = zip(*df_head_hunter['График'].apply(schedule))

# удаление ненужных столбцов
df_head_hunter = df_head_hunter.drop(['График', 'Занятость'], axis=1)
```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# считываем файл, для столбца "date" указываем тип данных datetime
df_currency = pd.read_csv('https://dl.dropboxusercontent.com/s/6gqwkk1fu86jdwu/ExchangeRates.csv?dl=0', delimiter=',')
df_currency['date'] = pd.to_datetime(df_currency['date'], dayfirst=True).dt.date

# устанавливаем тип данных для столбца "Обновление резюме" как datetime
df_head_hunter['Обновление резюме'] = pd.to_datetime(df_head_hunter['Обновление резюме'], dayfirst=True).dt.date

# функция по разделению данных из столбца "ЗП"
def desire_salary(money):
    money, curr = money.split(' ')
    money = float(money)
    return money, curr

# разделение данных, применение функции
df_head_hunter['Желаемая зарплата'], df_head_hunter['Валюта зп'] = zip(*df_head_hunter['ЗП'].apply(desire_salary))

# подготовительные словарь и список для исправления название валют
money_curr = {'бел.руб.': 'BYN', 'грн.': 'UAH', 'сум': 'UZS'}
bad_money_names = ['бел.руб.', 'грн.', 'сум']

# функция по исправлению названия валют
def currency_title(curr_name):
    if curr_name in bad_money_names:
        curr_name = money_curr[curr_name]
    return curr_name

# применение функции по изменению валют
df_head_hunter['Валюта зп'] = df_head_hunter['Валюта зп'].apply(currency_title)

# объединение датафреймов по названию валют и датам обновлению резюме/курса валют
df_head_hunter = df_head_hunter.merge(df_currency, how='left', left_on=['Валюта зп', 'Обновление резюме'], right_on=['currency','date'])

# заполнение пустых столбцов единицами
df_head_hunter['close'].fillna(1, inplace=True)
df_head_hunter['proportion'].fillna(1, inplace=True)
df_head_hunter['currency'].fillna('RUB', inplace=True)

# подсчет в рублях желаемой зарплаты
df_head_hunter['ЗП (руб)'] = df_head_hunter['Желаемая зарплата'] * df_head_hunter['close'] / df_head_hunter['proportion']

# удаляем лишние столбцы
df_head_hunter.drop(['Желаемая зарплата', 'Валюта зп', 'ЗП', 'close', 'date', 'currency', 'proportion', 'vol', 'per', 'time'], axis=1, inplace=True)
```
